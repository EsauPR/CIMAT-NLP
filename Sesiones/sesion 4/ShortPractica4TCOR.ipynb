```python
def sortFreqDict(freqdict):
    aux = [(freqdict[key], key) for key in freqdict]
    aux.sort()
    aux.reverse()
    return aux
```

```python
V = sortFreqDict(fdist)
V = V[:5000]
```

```python
dict_indices = dict()
cont = 0
for weight, word in V:
    dict_indices[word] = cont
    cont += 1   
```

### Bolsa de TÃ©rminos

```python
import numpy as np
def build_bow_tr(tr_txt, V, dict_indices):
    BOW = np.zeros((len(tr_txt),len(V)), dtype=int)
    cont_doc = 0
    for tr in tr_txt:
        fdist_doc = nltk.FreqDist(tr.split())
        for word in fdist_doc:
            if word in dict_indices:
                BOW[cont_doc, dict_indices[word]] = fdist_doc[word]
        cont_doc += 1
        
    return BOW
```

```python
import math

def compute_tcor(BOW, ndocs_out=0, y=None):
    #TCOR: (1 + math.log10(doc[term])) * math.log10(V_size/v_size)
    # Â¡Â¡Â¡TÃš CÃ“DIGO VA AQUÃ!!!
    TCOR = np.zeros((BOW.shape[1], BOW.shape[1]), dtype=np.float)
    print("DOR: Shape of the input matrix (BoT):", BOW.shape)
    print("DOR: Shape of the term-feat matrix:", TCOR.shape)

    size = BOW.shape[1]

    for k in range(size):
        if k % 100 == 0:
            print(k)
        # co-occurrences from term k with all the rest
        co_occur_k = BOW & np.tile(BOW[:,k], (size, 1)).T > 0

        # #(t_k, t_j)
        # tf_k[j] Number of documents where the term k and j co-ocurr
        cat_t_k = np.sum(co_occur_k, axis=0)
        zeros = cat_t_k == 0
        cat_t_k[zeros] = 1 # Conver zeros to ones to avoid log error
        tf_k = 1.0 + np.log(cat_t_k)
        tf_k[zeros] = 0 # Restore zeros

        # tao_tk[j] Count the terms that co-ocurrs with term_k in at least one document
        tao_tk = np.sum(np.any(co_occur_k, axis=0))
        
        TCOR[k] = tf_k * np.log(size / tao_tk)

        # for j in range(k, size):
        #     # #(t_k, t_j) Number of documents where the term k and j co-ocurr
        #     coor_tk = np.sum(((BOW[:, k] != 0) & (BOW[:,j] != 0)))
        #     tf_kj = 1.0 + math.log10(coor_tk) if coor_tk > 0 else 0
        #     TCOR[k][j] = TCOR[j][k] = tf_kj * math.log10(size / tao_tk)

    return TCOR
```

```python
from sklearn import preprocessing
TCOR_base = compute_tcor(BOW_tr)
TCOR_base = preprocessing.normalize(TCOR_base, norm='l2')
```

# Hay muchas formas de seleccionar las mejores caracterÃ­sticas (Tal vez despuÃ©s veamos alguna). De momento usamos Chi2 para seleccionar las 1000 mejores caracterÃ­sticas a partir de la BOW

```python
from sklearn.feature_selection import SelectKBest, chi2
feats = SelectKBest(chi2, k=1000)
feats.fit(BOW_tr, tr_y)
```

```python
best=feats.get_support(indices=True)
```

```python
dict_indice_invertido = {}
for w in dict_indices:
    dict_indice_invertido[dict_indices[w]] = w
```

```python
target_words = [dict_indice_invertido[index] for index in best]
words = [word for word in target_words if word in dict_indices] 
target_matrix = np.array([TCOR_base[dict_indices[word]] for word in words])
```

```python
from tsne import tsne
reduced_matrix = tsne(target_matrix, 2);
```
