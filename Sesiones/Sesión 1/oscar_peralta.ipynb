**Densidad L√©xica**

```
def lex_div(text):
    """ Longitud del texto / palabras √∫nicas """
    return len(text)/len(set(text))

print(lex_div(text1))
```

**Frecuencias**

```
from nltk import FreqDist

fdist1 = FreqDist(text1)
fdist1
```

```
# 50 elementos m√°s comunes 
vocabulary1 = fdist1.most_common(50)
list(vocabulary1)
```
